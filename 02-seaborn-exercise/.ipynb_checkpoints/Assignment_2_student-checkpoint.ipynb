## 2: Plot 12 most dangerous cities

```python
kill = kill.dropna()
```

```python
most_dangerous = kill.value_counts("city").head(12)
most_dangerous
```

```python
plt.figure(figsize=(8,5))

chart = sns.barplot(x=most_dangerous.index, y=most_dangerous.values)
chart.set(title="Most dangerous cities", xlabel=None)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

```python
# with seaborn barplot
```

```python
plt.figure(figsize=(8,5))

chart = sns.barplot(x=most_dangerous.index, y=most_dangerous.values)
chart.set(title="Most dangerous cities", xlabel="city", ylabel="count")
chart.set_xticklabels(chart.get_xticklabels(), rotation=60);
```

```python
# with seaborn countplot
```
